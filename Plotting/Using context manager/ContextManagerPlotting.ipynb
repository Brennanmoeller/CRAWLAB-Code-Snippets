```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x1 = np.linspace(0,5,501)
y1 = np.sin(2*np.pi * x1)

x2 = np.linspace(0,5,501)
y2 = np.sin(2.2*np.pi*x2)
```

### Defaults
Let's first plot using the matplotlib defaults.

*Note:* This plot formatting will very likely vary from the below when you plot it. I have set my defaults as close as possible to my desired style.

```python
plt.plot(x1, y1, linestyle="-", label=r'Data 1')
plt.plot(x2, y2, linestyle="--", label=r'Data 2')
```

```python
with plt.style.context(('CRAWLAB.mplstyle')):
    # Set the plot size - 3x2 aspect ratio is best
    fig = plt.figure(figsize=(6,4))

    # Remove the top and right splines. As of 01/11/15, we can't do this in the style file
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # Define the X and Y axis labels
    plt.xlabel('X label (units)', labelpad=5)
    plt.ylabel('Y label (units)', labelpad=10)

    # Plot the data and set the labels to be used in the legend
    plt.plot(x1, y1, linestyle="-", label=r'Data 1')
    plt.plot(x2, y2, linestyle="--", label=r'Data 2')
    # plot(x3, y3, linestyle="-.", label=r'Data 3')
    # plot(x4, y4, linestyle=":", label=r'Data 4')

    # Uncomment below and set limits if needed
    # plt.xlim(0,5)
    plt.ylim(-1,1.5)

    # Create the legend
    leg = plt.legend(loc='upper right', ncol = 2)

    # OPTIONAL : Adjust the page layout filling the page using the new tight_layout command
    # plt.tight_layout(pad=0.5)

    # save the figure as a high-res pdf in the current folder
#     plt.savefig('plot_from_Notebook.pdf')
    
    fig.set_size_inches(9,6) # Resize the figure for better display in the notebook
```

### XKCD style
How about an xkcd-style plot? It's built right into matplotlib.

```python
plt.rc('text', usetex=False)

with plt.xkcd(scale=1.25, length = 100, randomness = 2):

    # Set the plot size - 3x2 aspect ratio is best
    fig = plt.figure(figsize=(6,4))
    ax = plt.gca()
    plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

    # Change the axis units to xkcd font
    plt.setp(ax.get_ymajorticklabels(), fontsize=18)
    plt.setp(ax.get_xmajorticklabels(), fontsize=18)

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # Turn on the plot grid and set appropriate linestyle and color
    ax.grid(True,linestyle=':', color='0.75')
    ax.set_axisbelow(True)

    # Define the X and Y axis labels
    plt.xlabel('X label (units)', fontsize=22, labelpad=5)
    plt.ylabel('Y Label (units)', fontsize=22, labelpad=10)

    # Plot the data and set the labels to be used in the legend
    plt.plot(x1, y1, linestyle="-", label=r'Data 1')
    plt.plot(x2, y2, linestyle="--", label=r'Data 2')

    # uncomment below and set limits if needed
    # plt.xlim(0,5)
    plt.ylim(-1,1.25)

    # Create the legend, then fix the fontsize
    leg = plt.legend(loc='upper right', ncol = 2, fancybox=True)
    ltext  = leg.get_texts()
    plt.setp(ltext, fontsize=16)

    # Adjust the page layout filling the page using the new tight_layout command
    plt.tight_layout(pad=0.5)

    # save the figure as a high-res pdf in the current folder
    plt.savefig('plot_filename_xkcd.pdf')

    fig.set_size_inches(9,6) # Resize the figure for better display in the notebook

# return to using LaTeX
plt.rc('text', usetex=True)
```
